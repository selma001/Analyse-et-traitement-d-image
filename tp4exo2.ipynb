```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def calculate_laplacian_pixel_value(pixel_value, neighbors):
    laplacian_value = 4 * pixel_value - sum(neighbors)
    return laplacian_value
```

```python
def apply_laplacian_custom(image):  
    
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    laplacian_filter = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])
    laplacian = cv2.filter2D(gray, cv2.CV_64F, laplacian_filter)

    return laplacian
```

```python
image_path = "grayscale.png"
image = cv2.imread(image_path)

laplacian_custom = apply_laplacian_custom(image)

plt.imshow(laplacian_custom, cmap='gray')
plt.title('Laplacian (Custom)')
plt.show()
```

```python
def apply_edge_detection(laplacian_custom, threshold):
    # Create a binary image where edges are above the threshold
    edges = np.zeros_like(laplacian_custom)
    edges[laplacian_custom > threshold] = 255

    return edges
```

```python
threshold_value = 60

# Apply edge detection based on magnitude and threshold
edges = apply_edge_detection(laplacian_custom, threshold_value)

# Display the results
plt.imshow(edges, cmap="gray")
plt.title('Edge Detection')
plt.show()
```
