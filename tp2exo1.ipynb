```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('grayscale.png', 0)
img_np = np.array(image)
print(image)
```

```python
list_np = np.zeros(256, dtype=int)
rows, cols = img_np.shape

for i in range(rows):
    for j in range(cols):
        value = img_np[i, j]
        list_np[value] += 1
print(list_np)
```

```python
x = np.arange(list_np.size)
plt.bar(x,list_np, width= 1.0)

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Counts')
plt.title('Histogram of the image')

plt.show()
```

```python
img2 = img_np
for i in range(rows):
    for j in range(cols):
        if (img2[i, j] > 123):
            img2[i, j] = 255
        else:
            img2[i, j] = 0
plt.imshow(img2,cmap='gray')
```
